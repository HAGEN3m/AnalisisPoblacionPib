#### 3. Distribución de datos clave

```python
sns.histplot(df["Population"], kde=True)
sns.histplot(df["Gross Domestic Product (GDP)"], kde=True)
sns.histplot(df["Gross National Income(GNI) in USD"], kde=True)
```

Transformación logarítmica de variables económicas

Las variables como el Producto Interno Bruto (PIB) y el Ingreso Nacional Bruto (GNI) presentan un rango de valores considerablemente amplio, que abarca desde miles hasta billones de unidades monetarias. Esta alta dispersión genera una distribución marcadamente sesgada en escala lineal, lo que provoca que los datos de menor magnitud queden visualmente comprimidos, dificultando una interpretación adecuada del comportamiento general de la variable.  

Con el fin de reducir la asimetría de la distribución, facilitar su visualización y mejorar la interpretabilidad del análisis, se aplicará una transformación logarítmica en base 10 (log₁₀).


```python
plt.figure(figsize=(10, 6)) 
sns.histplot(np.log10(df["Population"]), kde=True, label = "Population", color = "skyblue")
sns.histplot(np.log10(df["Gross Domestic Product (GDP)"]), kde=True, label='GDP', color='orange')
sns.histplot(np.log10(df["Gross National Income(GNI) in USD"]), kde=True, label='GNI', color='green')

plt.title("Distribución Logarítmica de Variables Económicas")
plt.xlabel("Valor log₁₀")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

📊 Interpretación de la escala logarítmica en el eje X

El eje X del gráfico está expresado en escala logarítmica base 10 (log₁₀), lo que significa que cada número representa una potencia de 10. Esto permite representar de forma más clara valores económicos muy amplios expresados en dólares estadounidenses.

Ejemplos:
- 4 → 10⁴ = 10.000 USD
- 6 → 10⁶ = 1.000.000 USD
- 9 → 10⁹ = 1.000.000.000 USD
- 12 → 10¹² = 1.000.000.000.000 USD

El eje Y muestra la **frecuencia absoluta** de observaciones dentro de cada rango de valores en la escala logarítmica.


 #### 4. Relación entre población y PIB

```python
latest_year = df["Year"].max()
df_latest = df[df["Year"] == latest_year]

top = df_latest.sort_values("Population", ascending=False).head(10)["Country"].unique()

sns.scatterplot(data=df_latest[df_latest["Country"].isin(top)], x = "Population", y = "Gross Domestic Product (GDP)", hue = "Country", alpha = 0.6)
plt.title("Relación entre Población y PIB del año más reciente")

# Mover la leyenda a la derecha
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()  # Para que no se corte la leyenda
plt.show()
```

#### 5. Correlación entre variables numéricas

```python
cols_interes = ["Population", "Gross Domestic Product (GDP)", "Gross National Income(GNI) in USD"]
corr_subset = df[cols_interes].corr()

#Heatmap

plt.figure(figsize=(6,4))
sns.heatmap(corr_subset, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación entre variables clave")
plt.show()
```

 #### 6. Análisis temporal

```python
gdp_year = df.groupby("Year")["Gross Domestic Product (GDP)"].sum()
pop_year = df.groupby("Year")["Population"].sum()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24,10))

# Gráfico 1: PIB
ax1.plot(gdp_year.index, gdp_year.values, color = "blue")
ax1.set_title("Evolución del PIB global (en billones de USD)")
ax1.set_xlabel("Año")
ax1.set_ylabel("PIB (USD)")

# Gráfico 2: Población
ax2.plot(pop_year.index, pop_year.values, color = "red")
ax2.set_title("Evolución de la población global (en miles de millones)")
ax2.set_xlabel("Año")
ax2.set_ylabel("Población")

plt.tight_layout()
plt.show()
```
